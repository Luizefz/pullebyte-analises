```python
import pandas as pd
import matplotlib.pyplot as plt
from mplsoccer.pitch import Pitch
import mplcursors
import seaborn as sns

def load_data(file_path):
    return pd.read_parquet(file_path)

def filter_data(time_completo_df, club_id):
    return time_completo_df[time_completo_df['club_id'] == club_id]

time_completo_df = load_data(r'F:\Universidade\PISI3\pullebyte-analises\DataSet Project\merge-data-by-clubs\merge-time-completo.parquet')
```

```python
time_completo_df
```

```python
Barcelona = filter_data(time_completo_df, 131)
Barcelona
chutes_penalty_area =  Barcelona['penalties'].sum()
chutes_penalty_box  =  Barcelona['inside_area'].sum()
chutes_pitch_box    =  Barcelona['outside_areas'].sum()

# Configurações do campo com estilo específico
fig, ax = plt.subplots(figsize=(13, 9))
fig.set_facecolor('#0F1821')
ax.patch.set_facecolor('#0F1821')

# Criando o campo
pitch = Pitch(pitch_type='statsbomb', half=True, 
              pitch_color='#0F1821', line_color='#c7d5cc')
pitch.draw(ax=ax)

# Dimensões e posicionamento das áreas:
penalty_box = plt.Rectangle((102, 18), 18, 44, color='blue', alpha=0.3, label='Grande Área: 15 chutes')
pitch_box = plt.Rectangle((60, 0), 60, 80, color='green', alpha=0.1, label='Área Fora da Grande Penalidade: 25 chutes')
penalty_area = plt.Rectangle((102, 30), 6, 20, color='red', alpha=0.5, label='Área de Pênalti: 5 chutes')

# Adicionando as áreas ao campo
ax.add_patch(penalty_box)
ax.add_patch(pitch_box)
ax.add_patch(penalty_area)

# Adicionando labels para cada área
ax.text(115, 40, str(chutes_penalty_box), color='white', fontsize=15, ha='center')
ax.text(85, 40, str(chutes_pitch_box), color='white', fontsize=15, ha='center')
ax.text(105, 40, str(chutes_penalty_area), color='white', fontsize=15, ha='center')

# Criando tooltips
mplcursors.cursor(hover=True).connect(
    "add", lambda sel: sel.annotation.set_text(sel.artist.get_label())
)

# Ajustes finais
plt.gca().invert_yaxis()
plt.show()
```
